```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df.head()
```

```python
#1: To fill the missing values and to make the column integer column. These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place.

df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)

df.head()
```

```python
#2: To split 'From_To' into two separate columns and create temporary dataframe. 

temp_df = df.From_To.str.split('_', expand=True)

temp_df.columns = ['From', 'To']

temp_df.head()
```

```python
#3: To standardize the strings so that only the first letter of the city names is in uppercase

temp_df = temp_df.apply(lambda x: x.astype(str).str.title())
temp_df.head()
```

```python
#4: To delete the From_To column from df and attach the temporary DataFrame.

df = df.drop('From_To', axis=1)
df = df.join(temp_df)

df.head()
```

```python
#5: To expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df with delays.

delays = df['RecentDelays'].apply(pd.Series)
delays.columns = ['delays_{}'. format(n) for n in range(1, len(delays.columns)+1)]
df = df.drop('RecentDelays', axis =1).join(delays)

df.head()
```
